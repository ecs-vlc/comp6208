# Bias Variance Dilemma

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
#rc('text', usetex=True)
import math
from scipy.stats import norm
%matplotlib inline
```

```python
def genData(m,f):
    x = np.random.randn(m)
    y = np.vectorize(f)(x)
    return x, y

def fitfunc(x,y,p):
    pf = np.polyfit(x,y,p)
    return np.poly1d(pf)

def fit(m, p, f, show):
    x, y = genData(m, f)
    fhat = fitfunc(x, y, p)
    xs = np.linspace(3,-3)
    ys = np.vectorize(f)(xs)
    yps = fhat(xs)
    trainingError = sum((fhat(x)-y)**2)
    px = norm.pdf(xs)
    genError = np.dot(px,(yps-ys)**2)
    if show:
        plt.plot(xs,ys,"-r",xs,yps,"-b",x,y,"x")
        plt.plot([0,0],[-4,4], "-k", [-3,3],[0,0],"-k")
        plt.axis([-3,3,-3,3])
        text = "$E_l$ = {:.5f}, $E_g$ = {:.4f}".format(float(trainingError), float(genError))
        plt.text(-2.9,2,text)
        plt.xlabel("x")
        plt.ylabel("$\hat{f}\,(x)$")
        plt.show()
 
    return trainingError, genError

tE, gE = fit(10, 5, math.tanh, True)
print(tE, gE)
```

```python
def findMeanMachine(m, p, f, n=10000):
    meanpf = 0
    for i in range(n):
        x, y = genData(m, f)
        meanpf += np.polyfit(x,y,p)
    meanpf /= n
    return np.poly1d(meanpf)


meanMachine = findMeanMachine(11,5,math.tanh)

def showPoly(pf, f):
    xs = np.linspace(3,-3)
    ys = np.vectorize(f)(xs)
    yps = pf(xs)
    px = norm.pdf(xs)
    bias = np.dot(px,(yps-ys)**2)
    plt.plot(xs,ys,"-r",xs,yps)
    plt.plot([0,0],[-4,4], "-k", [-3,3],[0,0],"-k")
    plt.axis([-3,3,-3,3])
    text = "bias = {:.3f}".format(float(bias))
    plt.text(-2.9,2,text)
    plt.xlabel("x")
    plt.ylabel("$f_m(x)$")
    plt.show()
    return bias
 
bias = showPoly(meanMachine, math.tanh)

```

```python
def findVariance(m, p, f, n=10000):
    meanMachine = findMeanMachine(m,p,f,n)
    xs = np.linspace(3,-3)
    ys = np.vectorize(f)(xs)
    ymm = meanMachine(xs)
    px = norm.pdf(xs)
    bias = np.dot(px,(ymm-ys)**2)
    genError = 0
    trainError = 0
    variance = 0
    for i in range(n):
        x, y = genData(m, f)
        fhat = fitfunc(x,y,p)
        yps = fhat(xs)
        trainError += sum((fhat(x)-np.vectorize(f)(x))**2)
        genError += np.dot(px,(yps-ys)**2)
        variance += np.dot(px,(yps-ymm)**2)
    genError /= n
    variance /= n
    trainError /= n
    return bias, variance, genError, trainError

b,v,g,_ = findVariance(11, 5, math.tanh)
print("Bias = {:.3f} Variance = {:.3f} B+V = {:.3f} E_g = {:.3f}".format(b,v,b+v,g))
```

```python
bias = []
variance = []
genError = []
trainError = []
pp = range(1,10)
for p in pp:
    b,v,g,t = findVariance(11, p, math.tanh)
    bias.append(b)
    variance.append(v)
    genError.append(g)
    trainError.append(t)

pp = [p for p in pp]
plt.plot(pp, trainError, "-k", label="$E_t$")
plt.plot(pp, bias, "-b", label="bias")
plt.plot(pp, variance, "-r", label="variance")
plt.plot(pp, genError, "-g", label="$E_g$")
plt.xlabel("Polynomial, p")
plt.axis([1,9,0,100])
plt.legend()
plt.show()

plt.plot(pp, trainError, "-k", label="$E_t$")
plt.plot(pp, bias, "-b", label="bias")
plt.plot(pp, variance, "-r", label="variance")
plt.plot(pp, genError, "-g", label="$E_g$")
plt.xlabel("Polynomial, p")
plt.legend()
plt.axis([1,9,0,4])
plt.show()
```
